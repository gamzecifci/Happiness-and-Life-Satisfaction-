# 1. Life expectancy and life satisfaction

```python
fig, axes = plt.subplots(1,2, sharey=True, figsize=(15,7))
fig.tight_layout()


ax0 = axes[0] 
sns.boxplot(x='ladder_score', y='region', palette="rocket", ax=ax0, data=df)
ax0.set_ylabel('')

ax1 = axes[1]
sns.swarmplot(x='healthy_life', y='region', palette="rocket", ax=ax1, data=df)
ax1.set_ylabel('')
```

# 2.Corruption

```python
sns.lmplot( x="corruption", y="ladder_score", data=df, fit_reg=False, hue='region', legend=False, )

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

# 3.Social Support

```python
g = sns.jointplot(
    data=df,
    x="social_support", y="ladder_score", hue="region", kind="kde")
```

# 4.Economics

```python
region_score= df.groupby(['region']).agg({'ladder_score': 'mean', 'GDP': 'mean'}).sort_values('GDP')
region_score
```

```python
heat_map2 = df.groupby(['region']).agg({'GDP': 'mean'}).sort_values('GDP')
sns.heatmap(heat_map2,cmap='plasma',annot=True);
```
